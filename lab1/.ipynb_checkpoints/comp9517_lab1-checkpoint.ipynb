```python
import cv2
import matplotlib.pyplot as plt
```

# 1. Contrast Stretching

```python
cat_img = cv2.imread("Cat.png", -1)
plt.imshow(cat_img, cmap='gray', vmin=0, vmax=255)
plt.show()
```

Find the minimum & maximum pixel values

```python
c = min(map(min, cat_img))
d = max(map(max, cat_img))
```

Perform contrast stretching

```python
a = 0
b = 255
img_ansewer1 = (cat_img - c) * ((b-a)/(d-c)) + a
plt.imshow(img_ansewer1, cmap="gray", vmin=0, vmax=255)
plt.show()
```

# 2. Histogram Calculation

```python
# original image
row, col = cat_img.shape
gray_level = [0 for x in range(0, 256)]
for r in range(row):
    for c in range(col): 
        gray_level[cat_img[r,c]] += 1 

plt.bar(x=range(0,256), height = gray_level)
```

```python
# answer1 image
row, col = img_ansewer1.shape
gray_level = [0 for x in range(0, 256)]
for r in range(row):
    for c in range(col): 
        gray_level[int(img_ansewer1[r,c])] += 1 

plt.bar(x=range(0,256), height = gray_level)
```

# 3. Image Smoothing

```python
# original moon
moon_img = cv2.imread("Moon.png", -1)
plt.imshow(moon_img, cmap='gray')
plt.title('Original')
plt.show()
```

mean (uniform) filter to smooth

```python
#Creating the kernel
k = 9
#kernel_mean = [[1] * k for i in range(k)]
row = moon_img.shape[0]
col = moon_img.shape[1]
```

```python
mean_moon = [[0] * row for i in range(col)]
for i in range(4, row-4):
    for j in range(4, col-4):
        mat = moon_img[i:i+k, j:j+k]
        mean_moon[i][j] = (mat.sum())/(k^2)

plt.imshow(mean_moon, cmap='gray')
plt.title('Mean')
plt.show()
```

median filter to smooth

```python
#Creating the kernel
k = 3
median_moon = [[0] * row for i in range(col)]
for i in range(1, row-2):
    for j in range(1, col-2):
        # sort
        mat = moon_img[i:i+k, j:j+k]
        new = list(i for j in mat for i in j)
        new = sorted(new)
        median_moon[i][j] = new[int(len(new)/2)]

plt.imshow(median_moon, cmap='gray')
plt.title('Mediam')
plt.show()
```

The best filter is median filter and kernel size is 3*3 for this image.


# 4. Edge Detection (Marked)

```python
import math
img_CT = cv2.imread("CT.png",-1)
plt.imshow(img_CT, cmap='gray')
plt.show()
```

```python
sobel_x = [[1, 0, -1],[2, 0, -2],[1, 0, -1]]
sobel_y = [[1, 2, 1],[0, 0, 0],[-1, -2, -1]]
row, col= img_CT.shape

new_CT = [[0] * col for i in range(row)]

for i in range(1, row-2):
    for j in range(1, col-2):
        dx = (img_CT[i:i+3,j:j+3] * sobel_x).sum()
        dy = (img_CT[i:i+3,j:j+3] * sobel_y).sum()
        pixel = math.sqrt(math.pow(dx,2) + math.pow(dy,2))
        if pixel > 160:
            new_CT[i][j] = 255 
        else:
            new_CT[i][j] = pixel

plt.imshow(new_CT, cmap='gray')
plt.title("Output")
plt.show()
```
